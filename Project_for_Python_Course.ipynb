```python
#Uploading Datasets: Titanic: Machine Learning from Disaster
#And read for further analysis
import pandas as pd

df = pd.read_csv('Titanic Machine Learning from Disaster.csv')
```

```python
#1:Finding youngest and oldest survivors from the dataset
survivors = df[df['Survived'] == 1]
youngest_survivor = survivors.loc[survivors['Age'].idxmin()]
oldest_survivor = survivors.loc[survivors['Age'].idxmax()]

print("Youngest Survivor:\n", youngest_survivor)
print("\nOldest Survivor:\n", oldest_survivor)
```

```python
#2:A bar chart showing the distribution of passengers by Name
import matplotlib.pyplot as plt

df['Name'].value_counts().plot(kind='bar')
plt.title('Distribution of Passengers by Name')
plt.xlabel('Name')
plt.ylabel('Count')
plt.show()
```

```python
#3:Survival rate for passengers with more than 2 family members onboard
df['FamilySize'] = df['SibSp'] + df['Parch']
large_families = df[df['FamilySize'] > 2]
survival_rate = large_families['Survived'].mean()

print(f"Survival rate for passengers with more than 2 family members: {survival_rate:.2%}")
```

```python
#4:Replacing missing deck values with the mode of the Cabin column
deck_mode = df['Cabin'].mode()[0]
df['Cabin'].fillna(deck_mode, inplace=True)

print("Missing deck values replaced with mode:", deck_mode)
```

```python
#4:Replacing missing deck values with the mode of the Age column
deck_mode = df['Age'].mode()[0]
df['Age'].fillna(deck_mode, inplace=True)

print("Missing deck values replaced with mode:", deck_mode)
```

```python
#5:A violin plot of fare distribution for survivors and non-survivors by class
import seaborn as sns
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=df, split=True, inner='quartile', palette='muted')
plt.title('Fare Distribution by Class and Survival Status')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
#6:Total number of male passengers who embarked from Queenstown
male_queenstown = df[(df['Sex'] == 'male') & (df['Embarked'] == 'Q')].shape[0]
print(f"Total number of male passengers who embarked from Queenstown: {male_queenstown}")
```
